```python
import numpy as np
import cv2 as cv
class1=[347,378,355,385,341,346,410,345,388,370,379,355]
def hisEqulColor2(img):
    
    ycrcb = cv.cvtColor(img, cv.COLOR_BGR2YCR_CB)
    
    channels = cv.split(ycrcb)
    
    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    clahe.apply(channels[0], channels[0])
    
    cv.merge(channels, ycrcb)
    cv.cvtColor(ycrcb, cv.COLOR_YCR_CB2BGR, img)
    return img

# for i in range(1,3,1):
#     c=class1[i-1]
#     for j in range(c):
#         print('img/label'+str(i)+'-'+str(j+1)+'.jpg')
#         #img = cv.imread('img/label1-1.jpg')
#         #img2 = img.copy()
#         #res2 = hisEqulColor2(img2)

#         #cv.imwrite('1/mosaic_2_deHaze_hist.jpg', res2)
#print('img/labej=l1-'+str(1)+'.jpg')
for i in range(1,2,1):
    c=class1[i-1]
    for j in range(1,c,1):
        img = cv.imread('img/label'+str(i)+'-'+str(j)+'.jpg')
        img2 = img.copy()
        res2 = hisEqulColor2(img2)
        cv.imwrite('CLAHE/label'+str(i)+'-'+str(j)+'.jpg', res2)
        print('img/label'+str(i)+'-'+str(j)+'.jpg')
```
